```python
import pandas as pd
import json
```

```python
def joinKeywordResultsToCompanyIDs(company_names_df,keywords_results_df):
    keywords_results_df['CompanyName'] = keywords_results_df['name']
    keywords_results_df = keywords_results_df.drop('name',axis=1)
    merged_df = pd.merge(company_names_df,keywords_results_df,how='left',on='CompanyName')
    
    
    return merged_df
```

```python
def createKeywordsJSON(company_name_list_df):
    
    def cleanupText(text,capitals=True,remove_punctuation=False):
        if remove_punctuation:
            from string import punctuation
            text = ' '.join(char for char in text if char not in punctuation)
        if capitals:
            text = text.lower()
        return text


    keywords_df = company_name_list_df['CompanyName'].apply(cleanupText)
    
 
            
        
    return keywords_df.to_frame().rename(columns={'CompanyName':'keywords'})

    
```
